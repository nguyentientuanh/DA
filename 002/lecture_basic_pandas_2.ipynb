```python
# XỬ LÝ DỮ LIỆU TRÊN PANDAS

# Pandas là gì? thư viện trên Python - xử lý dữ liệu, phân tích dữ liệu

# Thành phần trong Pandas:
  # 1. DataFrame: Bảng, có dòng, có cột => dữ liệu 2 chiều
  # 2. Series: cột trong bảng => dữ liệu 1 chiều

# Tạo bảng thủ công
  # Tạo bảng rỗng: pd.DataFrame()
  # Tạo bảng từ danh sách: pd.DataFrame(list)
    # danh sách: list = [value1, value2, ...]
  # Tạo bảng từ tuple: pd.DataFrame(tuple)
    # tuple: tuple=[('Anh', 20),('Nam', 21)]
  # Tạo bảng từ dictionary: pd.DataFrame(dictionary)
    # dictionary: dictionary = {key1:value1, key2:value2, ...}

# Đọc dữ liệu:
  # pd.read_csv('file.csv')
  # json
  # excel
  # ...

# Xem dữ liệu: df
  # Xem qua:
    # dòng đầu: df.head()
    # dòng cuối: df.tail()
    # ngẫu nhiên: df.sample(5)
  # Xem kỹ
    # Hình thức
      # Hình dạng - có bao nhiêu dòng bao nhiêu cột: df.shape
      # Thông tin - Định dạng dữ liệu của từng cột: df.info()
    # Nội dung: tổng quát thống kê (Mô tả): df.describe()

# Truy cập dữ liệu
  # Gián tiếp (không biết tên, gọi bằng cách đếm số): df.iloc[dòng, cột]
  # Trực tiếp: df.loc[dòng, cột]

# Lọc dữ liệu
  # Lọc cột: age, gender, id => df[['age', 'gender', 'id']]
  # Lọc dựa trên điều kiện: df[điều kiện] => df[df['age'] <= 30]
    # VÀ: df[(điều kiện 1) & (điều kiện 2)]
    # HOẶC: df[(điều kiện 1) | (điều kiện 2)]

# Xử lý dữ liệu
  # Hình thức
    # Đổi tên cột
    # Tạo cột mới
    # Tạo dòng mới
    # Xoá cột
    # Xoá dòng
  # Nội dung:
    # Chỉnh sửa
```

# Cài đặt

```python
import pandas as pd
```

```python
df = pd.read_csv('FoodPrice_in_Turkey.csv')
```

```python
df.head()
```

# Xử lý dữ liệu

## Chỉnh sửa cấu trúc dữ liệu

### Đổi tên cột

```python
# Hàm để đổi tên cột: .rename()
# Áp dụng hàm này lên bảng dữ liệu: df.rename()
# Chỉ định muốn đổi tên cột nào: Place, ProductName
# Chỉ định muốn đổi tên thành gì: Địa điểm, Tên SP
# Code để chỉ định: columns={'Place':'Địa điểm', 'ProductName':'Tên SP'}

# Gộp tất cả: df.rename(columns={'Place':'Địa điểm', 'ProductName':'Tên SP'})
```

```python
# Cách 1
df = df.rename(columns={'Place':'Địa điểm', 'ProductName':'Tên SP'})
df.head()
```

```python
# Cách 2
df.rename(columns={'Place':'Địa điểm', 'ProductName':'Tên SP'}, inplace=True)
```

### Thêm cột mới

```python
# Cột mới
df['cột mới'] = 'mới'
df.head()
```

```python
# tạo cột 'giảm giá' có giá trị '15%'
df['Giảm giá'] = '15%'
df.head()
```

```python
# Tạo cột mới dựa trên cột sẵn có
# Weight, Height => BMI

# Tạo cột mới: 'giá mới' = giá ban đầu - 10/100*giá ban đầu
# -> df['giá mới'] = df['Price'] - df['Price']*10/100
```

```python
df['giá mới'] = df['Price']*(1-10/100)
df.head()
```

### Bỏ cột

```python
# Bỏ 'cột mới', 'giá mới', 'Giảm giá'
df1 = df[['Địa điểm',	'ProductId',	'Tên SP',	'UmId',	'UmName',	'Month',	'Year',	'Price']]
df1.head()
```

```python
# Cách 2: thực sự bỏ
# .drop
# df = df.drop(['cột mới', 'giá mới', 'Giảm giá'], axis=1)
df.drop(['cột mới', 'giá mới', 'Giảm giá'], axis=1, inplace=True)
```

```python
df
```

```python
# Bỏ cột 'Month', 'Year'
df = df.drop(['Month', 'Year'], axis=1)
df.head()
```

### Thêm dòng

```python
# .append()

# Địa điểm: 'NA'
# ProductId: 'RR'
# Tên SP: 'Rice'
# UmId: 10
# UmName: 'KG'
# Price: 84.3785

# Pandas 2.0 bỏ hàm append()
```

```python
df.head()
```

```python
df = df.append(pd.Series({'Địa điểm':'NA',
                'ProductId': 'RR',
                'Tên SP': 'Rice',
                'UmId': 10,
                'UmName': 'KG',
                'Price': 84.3785}),
               ignore_index=True)

df.tail()
```

### Xoá dòng

```python
# Lọc dữ liệu
```

```python
df.head()
```

```python
# .drop
df.drop(1, axis=0, inplace=True)
```

```python
df.reset_index(drop=True)
```

```python
# bỏ
# .drop
# bỏ cột: df.drop(cột, axis=1)
# bỏ dòng: df.drop(dòng, axis=0)
```

## Sửa dữ liệu: thay đổi nội dung

```python
# .replace()
# df.replace(giá trị cũ, giá trị mới, inplace=True)
```

```python
# Thay số 5 bằng số 10 trong toàn bộ tập dữ liệu
df.replace(5,10,inplace=True)
df
```

```python
# Thay mã sản phẩm từ giá trị 52 thành RR trong toàn bộ dữ liệu
df.replace(52, 'RR', inplace=True)
df
```

```python
df = df['ProductId'].replace(52, 'RR')
```

```python
df['ProductId'].replace(52, 'RR', inplace=True)
```

```python
# Thay ở cột UmName: KG -> kg
df['UmName'].replace('KG', 'kg', inplace=True)
```
