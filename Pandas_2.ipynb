## Phan tich bảng chéo Pivot Table



TH1: Tạo bảng đầy đủ các thông tin cần phân tích theo hàng, cột

```python
df.pivot_table(values='Price',index='ProductName',columns='Year',aggfunc ='mean')
```

TH2: Tạo bảng với các thông tin cần phân tích được tổ chức theo cột

```python
df.pivot_table(values='Price', columns='Year',aggfunc ='max')
```

TH3: Tạo bảng với các thông tin cần phân tích được tổ chức theo dòng

```python
df.pivot_table(values='Price', index='ProductName',aggfunc ='max')
```

TH4: Tạo bảng với các thông tin tính toán theo nhiều hàm của cùng một giá trị values

```python
df.pivot_table(values='Price', index='Month', columns='Place', aggfunc = {min, max})
```

TH5: Tạo bảng với các thông tin tính toán theo nhiều giá trị values

```python
import numpy as np
```

```python
df.pivot_table(values=['Price','Month'], index='ProductName', columns='Place', aggfunc ={'Price': np.sum,'Month': np.mean})
```

5.Với mỗi châu lục hãy tính tổng GDP; trung bình cộng GDP. Hợp nhất 2 bảng này thành một bảng duy nhất gồm 3 thông tin: Tên châu lục; Tổng GDP; TBC GDP.

```python
df_mean=df.groupby('Continent').agg({'GDP (millions of US$)':'mean'})
df_mean
```

```python
df_sum=df.groupby('Continent').agg({'GDP (millions of US$)':'sum'})
df_sum
```

```python
dfconcat=pd.concat([df_mean,df_sum],axis=1)
dfconcat
```
