## Truy cập dữ liệu

```python
# VỊ TRÍ
# Gọi theo số thứ tự của vị trí: integer location => df.iloc[3,4]
# Gọi theo tên: location => df.loc['Vietnam', 'Birth rate']
```

### iloc

```python
# 0, 1, 2, 3, ...
```

```python
# Truy cập dòng có chỉ số 3 của dữ liệu
df.iloc[3]
```

```python
# Truy cập các dòng từ vị trí 3 đến vị trí 9: 3:9
df.iloc[3:9]
```

```python
# Truy cập các dòng rời rạc: 3, 8, 9
df.iloc[[3, 8, 9]]
```

```python
# Truy cập cột 4 (0, 1, 2, 3, 4)
df.iloc[:, 4]
```

```python
# Truy cập cột từ cột 3 đến cột 7
df.iloc[:, 3:8]
```

```python
# Truy cập từ dòng 3 đến dòng 5, cột 5 đến cột 7 của dữ liệu
df.iloc[3:6, 3:8]
```

```python
# integer location -> iloc
# df.iloc[dòng, cột]
# df.iloc[dòng]
  # df.iloc[dòng1:dòng3]
  # df.iloc[[1, 4, 5]]
# df.iloc[:, cột]
```

### loc

```python
student_dict = {'Name': ['Ngân', 'Đạt', 'Hưng'],
                'Age': [20, 21, 19],
                'Marks': [85.10, 77.80, 91.54]}

# create DataFrame from dict
student_df = pd.DataFrame(student_dict)
student_df
```

```python
student_df = student_df.set_index('Name')
student_df
```

```python
student_df.loc['Hưng', 'Age']
```

```python
# điểm thi Marks của bạn Đạt
student_df.loc['Đạt','Marks']
```

#### Lọc dữ liệu

```python
# df[điều kiện]
# df[df['age'] <= 30]
# df[(df['age'] <= 30) & (df['gender']=='Female')]
# df[(df['age'] <= 30) | (df['gender']=='Female')]
```

```python
# Lọc dữ liệu: chỉ lấy dòng ProductName là Rice - Retail
df[df['ProductName']=='Rice - Retail']
```

```python
# ProductName là Rice - Retail VÀ Year là 2019
df[(df['ProductName'] == 'Rice - Retail') & (df['Year'] == 2019)]
```

```python
# XỬ LÝ DỮ LIỆU DÙNG PANDAS

# Pandas: thư viện trong Python -> chuyên dùng để phân tích dữ liệu
# Thành phần:
  # DataFrame: bảng có dòng có cột => dữ liệu 2 chiều
  # Series: mỗi cột trong bảng => dữ liệu 1 chiều

# Tạo bảng thủ công
  # Tạo bảng rỗng: pd.DataFrame()
  # Tạo bảng từ danh sách: pd.DataFrame(list)
    # list = [value1, value2, ...]
  # Tạo bảng từ tuple: pd.DataFrame(tuple)
    # tuple = [('Tòng', 19), ('Nam', 18), ...]
  # Tạo bảng từ dictionary: pd.DataFrame(dictionary)
    # dictionary = {key:value, key2:value2, ...}

# Đọc dữ liệu
  # pd.read_csv('file.csv')
  # pd.read_excel('file.xlsx')
  # pd.read_json(...)

# Xem dữ liệu
  # Xem qua
    # 5 dòng đầu: df.head()
    # 5 dòng cuối: df.tail()
    # 5 dòng bất kỳ: df.sample(5)
  # Xem kỹ
    # Hình thức
      # bao nhiêu dòng bao nhiêu cột: df.shape
      # định dạng dữ liệu của từng cột: df.info()
    # Nội dung
      # df.describe()

# Truy cập dữ liệu
  # df.iloc[hàng, cột] (integer location: truy cập dữ liệu bằng cách đánh số)
  # df.loc[hàng, cột] (location: truy cập dữ liệu bằng cách gọi tên)

# Chọn dữ liệu
  # Chọn cột: df[[cot1, cot2, cot4, ...]]
  # Chọn dòng (Lọc dữ liệu ): df[điều kiện]
    # và: df[(điều kiện 1) & (điều kiện 2)]
    # hoặc: df[(điều kiện 1) | (điều kiện 2)]
```
