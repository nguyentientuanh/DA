# Thêm cột

## Thêm cột số

```python
# Tạo cột 'rating' tính điểm cho mỗi cửa hàng dựa vào thông tin các cột rating_bad, rating_good, rating_normal
# theo công thức sau: rating = rating_good *2 + rating_normal - rating_bad*3
df['rating'] = df['rating_good']*2 + df['rating_normal'] - df['rating_bad']*3
```

```python
df.head()
```

## Thêm cột chữ

```python
df[['join_month', 'join_day', 'join_year']].info()
```

```python
# Chuyển đổi định dạng dữ liệu: .astype('dạng mình muốn')
```

```python
# ghép 3 cột join_month, join_day, join_year thành cột mới có tên 'date' nhận giá trị có dạng: "join_month join_day,join_year"
# April 30, 2024 -> chữ
# chữ + chữ -> tạo ra chữ mới
# số + số -> tạo ra giá trị mới
# chữ + số -> lỗi
df['date'] = df['join_month'] + " " + df['join_day'].astype('str') + ", " + df['join_year'].astype('str')
df[['join_month', 'join_day', 'join_year', 'date']]
```

## Thêm cột điều kiện

### 1 điều kiện

```python
df['new'] = df['join_year'] == 2021
```

```python
df[['join_year', 'new']]
```

### 2 điều kiện

```python
# Thêm cột rate có giá trị
  # good nếu rating_good >= 50000,
  # bad trong trường hợp còn lại

import numpy as np
df['rate']=np.where(df['rating_good'] >= 50000, 'good', 'bad')
```

```python
df[['rate', 'rating_good']]
```

### Nhiều điều kiện

```python
# Thêm cột flag tặng cờ cho các cửa hàng
# flag nhận các giá trị như sau:
  # blue khi rating_good >= 30000 và rating_bad <= 100
  # yellow khi 10000 <= rating_good < 30000 và 100 < rating_bad <= 1000
  # red khi rating_good < 10000
  # black đối với các trường hợp còn lại
```

```python
# blue
  # (df['rating_good'] >= 30000) & (df['rating_bad] <= 100)

# yellow
  # (df['rating_good'] >= 10000) &
  # (df['rating_good'] < 30000) &
  # (df['rating_bad'] > 100) &
  # (df['rating_bad'] <= 1000)

# red
  # df['rating_good'] < 10000

# black
```

```python
# 1. cho tất cả nhóm điều kiện vào 1 danh sách điều kiện
conditions = [(df['rating_good'] >= 30000) & (df['rating_bad'] <= 100),
              (df['rating_good'] >= 10000) & (df['rating_good'] < 30000) & (df['rating_bad'] > 100) & (df['rating_bad'] <= 1000),
              (df['rating_good'] < 10000)]

# 2. cho tất cả giá trị vào 1 danh sách
values = ['blue', 'yellow', 'red']

# 3. gán nhóm điều kiện với giá trị tương ứng dùng hàm np.select
df['flag'] = np.select(conditions, values, default='black')
```

```python
df[['rating_good', 'rating_bad', 'flag']]
```
